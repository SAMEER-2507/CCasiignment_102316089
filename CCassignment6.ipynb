Q1. Ask the user to enter a value (e.g., M) for a mathematical function.  Generate x values from -10 to 10 using np.linspace().  Compute y values for: o y= M⋅x 2 o y=M⋅sin(x)  Plot both functions in a single figure: o Use different colors and line styles. o Add legend, grid, and title.

```python
from cProfile import label
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
M=float(input("enter value for M"))
x=np.linspace(-10,10,100)
y1=M*x**2
y2=M*np.sin(x)
plt.Figure(figsize=(10,10))
plt.plot(x,y1,'r--',label='y=M*x^2')
plt.plot(x,y2,'b-',label='y=M*sin(x)')
plt.legend()
plt.grid()
plt.title("mathematical functions")
plt.show()
```

Q.2 Create a dataset of five subjects and scores.  Convert it into a Pandas DataFrame.  Plot the scores using a Seaborn bar plot with: o Different colors for each bar. o Annotations on top of each bar. o Title, axis labels, and grid.

```python
subjects=['math','sci','english','sst','hindi']
scores=[89,79,97,99,93]
df=pd.DataFrame({'subject':subjects,'score':scores})
plt.Figure(figsize=(10,10))
ax = sns.barplot(x='subject',y='score',data=df,palette='deep')

for i,score in enumerate(scores):
    ax.text(i,score+1,str(score),ha='center',fontsize=12,fontweight='bold')
plt.title("subject Score")
plt.xlabel("subjects")
plt.ylabel("scores")
plt.grid(axis='y')
plt.show()
```

Q.3 Question Explanation: Use the roll number as a seed for random number generation (np.random.seed()). Generate 50 random values using np.random.randn(), which creates normally distributed values. Create a 2×2 subplot layout with: A line plot showing the cumulative sum of the dataset. A scatter plot with random noise. Customize with titles, axis labels, and grids.

```python
rollno=int(input("enter rollno"))
np.random.seed(rollno)
data=np.random.randn(50)
fig,axs=plt.subplots(2,2,figsize=(10,8))
axs[0,0].plot(np.cumsum(data),'g-',label='cumulative sum')
axs[0,0].set_title('commulative sum plot')
axs[0,0].grid()
axs[0,1].scatter(range(len(data)),data,c='r',label="random noise")

axs[0,1].set_title('scatter  plot with random noise')
axs[0,1].grid()

plt.tight_layout()
plt.show()
```

Q.4 Apply ‘seaborn’ library to do the following.

1.Read Total profit of all months and show it using a line plot.

2.Read all product sales data and show it using a multiline plot.

3.Plot bar chart for all the features/attributes.

```python
df_sales = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")

plt.figure(figsize=(8, 5))
plt.plot(df_sales['month_number'], df_sales['total_profit'], marker='o', linestyle='-', color='b', label='Total Profit')
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Total Profit Over Months")
plt.grid()
plt.legend()
plt.show()
```

```python
from matplotlib.lines import lineStyles
from pyparsing import lineStart


plt.Figure(figsize=(10,10))
for column in df_sales.columns[1:-1]:
    plt.plot(df_sales['month_number'],df_sales[column],marker='o',linestyle='-',label=column)

plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.title("Product Sales Over Months")
plt.legend()
plt.grid()
plt.show()
```

```python
from matplotlib.lines import lineStyles
from pyparsing import lineStart


plt.Figure(figsize=(10,10))
df_sales.set_index('month_number').plot(kind='bar',figsize=(12,6))

plt.xlabel("Month Number")
plt.ylabel("Sales figure")
plt.title("sales data of all products")
plt.legend()
plt.grid()
plt.show()
```
